## Problem 3:  Gradient Descent and the learning rate

By modifying the learning rate below, show how the convergence takes longer or doesn't converge at all.
Can you explain in words or math why this is?

````python
from numpy import *
import pylab as plt 
# y = mx + b
# m is slope, b is y-intercept
def compute_error_for_line_given_points(b, m, points):
    totalError = 0
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        totalError += (y - (m * x + b)) ** 2
    return totalError / float(len(points))

def step_gradient(b_current, m_current, points, learningRate):
    b_gradient = 0
    m_gradient = 0
    N = float(len(points))
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        b_gradient += -(2/N) * (y - ((m_current * x) + b_current))
        m_gradient += -(2/N) * x * (y - ((m_current * x) + b_current))
    new_b = b_current - (learningRate * b_gradient)
    new_m = m_current - (learningRate * m_gradient)
    return [new_b, new_m]

def gradient_descent_runner(points, starting_b, starting_m, learning_rate, num_iterations):
    b = starting_b
    m = starting_m
    for i in range(num_iterations):
        b, m = step_gradient(b, m, array(points), learning_rate)
    return [b, m]

def run(num_iterations):
    points = genfromtxt("data.csv", delimiter=",")
    learning_rate = 0.0001
    initial_b = 0 # initial y-intercept guess
    initial_m = 0 # initial slope guess
    num_iterations = num_iterations
    print "Starting gradient descent at b = {0}, m = {1}, error = {2}".format(initial_b, initial_m, compute_error_for_line_given_points(initial_b, initial_m, points))
    print "Running..."
    [b, m] = gradient_descent_runner(points, initial_b, initial_m, learning_rate, num_iterations)
    print "After {0} iterations b = {1}, m = {2}, error = {3}".format(num_iterations, b, m, compute_error_for_line_given_points(b, m, points))
    for i in range(0,len(points)):
        plt.scatter(points[i,0],points[i,1])
        plt.scatter(points[i,0],m*points[i,0]+b,color='r')
    plt.show()

run(100)
````

````python
# Use this to visually discuss convergence rate based on learning rate

for num in range(0,10):
    run(num)
    plt.show()
````

````python
def run_difLR(learning_rate,num_iterations):
    # Running under different learning rates
    points = genfromtxt("data.csv", delimiter=",")
    initial_b = 0 # initial y-intercept guess
    initial_m = 0 # initial slope guess
    num_iterations = num_iterations
    [b, m] = gradient_descent_runner(points, initial_b, initial_m, learning_rate, num_iterations)
    error = compute_error_for_line_given_points(b, m, points)
    return error
learning_rate = logspace(-6,-3.4,20)
error_50 = 0*learning_rate
error_100 = 0*learning_rate
for i in range (len(learning_rate)):
    error_50[i] = run_difLR(learning_rate[i],50)
plt.plot(learning_rate,error_50)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Learning rate')
plt.ylabel('Error after 50 iterations')
plt.show()

for i in range (len(learning_rate)):
    error_100[i] = run_difLR(learning_rate[i],100)
plt.plot(learning_rate,error_100)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Learning rate')
plt.ylabel('Error after 100 iterations')
plt.show()

# When the learning rates are set too small, it will take more iterations to achieve the mininum error. 
# When the learning rates are set too big, it will never achieve the mininum error.
````

### Probelm 4.1. Lasso Regression

a) Use $L^1$ regularization on the scaled dataset (but original, with no new features), to the paramter
which yields the best $R^2$. Plot your result.

**Hint:** Take as your range of alpha:
`alphas = np.logspace(-3,0.5,30)`


b) Using the optimal constant, plot the feature coefficients - which one seems the least significant?


c) Now repeat the above, but take an average off 5 folds using cross validation. Do a boxplot of the coefficients

````python
# a
import pandas as pd
from matplotlib import pylab as plt
import seaborn
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import scale
import numpy.random as nprnd
import numpy as np
import random
import json
pd.set_option('display.max_columns', 500)
%matplotlib inline

df = pd.read_csv('http://www-bcf.usc.edu/~gareth/ISL/Advertising.csv', index_col=0)
y = df.iloc[:]['Sales']
X = df
del X['Sales']
X.head()
#Normalize the datasets
X_scaled = scale(X)

size = 200
X_train = X_scaled[0:int(size*0.8)]
X_test = X_scaled[int(size*0.8):]

y_train = y[0:int(size*0.8)]
y_test = y[int(size*0.8):]

alphas = np.logspace(-3,0.5,30)

scores = []

for alpha in alphas:
    regr = Lasso(alpha=alpha)

    # Train the model using the training sets
    regr.fit(X_train, y_train)
    scores.append(regr.score(X_test,y_test))
    
plt.plot(alphas,scores)
plt.xlabel('Alpha')
plt.ylabel('Scores')
plt.show()

#b 
#The coefficient of the newspaper is the least significant.

best_alpha = np.argmax(scores)
regr = Lasso(alpha=alphas[best_alpha])

regr.fit(X_train, y_train)
print('Coefficients: \n', regr.coef_)
print("Residual sum of squares: %.2f"
      % np.mean((regr.predict(X_test) - y_test) ** 2))
print('Variance score: %.2f' % regr.score(X_test, y_test))

#c

X_train_1 = X_scaled[0:int(size*0.8)]
X_test_1 = X_scaled[int(size*0.8):]
y_train_1 = y[0:int(size*0.8)]
y_test_1 = y[int(size*0.8):]

regr = Lasso(alpha=alphas[best_alpha])
regr.fit(X_train_1, y_train_1)
coefficents = regr.coef_

X_train_2 = np.vstack((X_scaled[0:int(size*0.6)],X_scaled[int(size*0.8):]))
X_test_2 = X_scaled[int(size*0.6):int(size*0.8)]
y_train_2 = np.append(y[0:int(size*0.6)],y[int(size*0.8):],axis = 0)
y_test_2 = y[int(size*0.6):int(size*0.8)]
regr = Lasso(alpha=alphas[best_alpha])
regr.fit(X_train_2, y_train_2)
coefficents = np.vstack((coefficents,regr.coef_))
 
X_train_3 = np.vstack((X_scaled[0:int(size*0.4)],X_scaled[int(size*0.6):]))
X_test_3 = X_scaled[int(size*0.4):int(size*0.6)]
y_train_3 = np.append(y[0:int(size*0.4)],y[int(size*0.6):],axis = 0)
y_test_3 = y[int(size*0.4):int(size*0.6)]
regr = Lasso(alpha=alphas[best_alpha])
regr.fit(X_train_3, y_train_3)
coefficents = np.vstack((coefficents,regr.coef_))

X_train_4 = np.vstack((X_scaled[0:int(size*0.2)],X_scaled[int(size*0.4):]))
X_test_4 = X_scaled[int(size*0.2):int(size*0.4)]
y_train_4 = np.append(y[0:int(size*0.2)],y[int(size*0.4):],axis = 0)
y_test_4 = y[int(size*0.2):int(size*0.4)]
regr = Lasso(alpha=alphas[best_alpha])
regr.fit(X_train_4, y_train_4)
coefficents = np.vstack((coefficents,regr.coef_))

X_train_5 = X_scaled[int(size*0.2):]
X_test_5 = X_scaled[0:int(size*0.2)]
y_train_5 = y[int(size*0.2):]
y_test_5 = y[0:int(size*0.2)]
regr = Lasso(alpha=alphas[best_alpha])
regr.fit(X_train_5, y_train_5)
coefficents = np.vstack((coefficents,regr.coef_))

plt.boxplot(coefficents)
plt.ylabel('Coefficients')
plt.xticks([1, 2, 3], ['TV', 'Radio', 'Newspaper'])
plt.show()  
````
